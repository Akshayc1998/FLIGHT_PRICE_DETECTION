# pie chart

```python
B2B_check=sushma['B2B'].value_counts()
plt.pie(B2B_check,labels=B2B_check.index,autopct="%2.1f%%",shadow='True',pctdistance=1.8,labeldistance=1.2)
plt.show()
```

```python
sushma['B2B'].value_counts()
```

```python
# Checking B2B Data  by using pie chart 
full =sushma['Fulfilment'].value_counts()

#  Plot the pie chart
plt.pie(full,labels=full.index,autopct="%1.2f%%",radius=2,frame=True)
#plt.axis('equal')
plt.show()
```

```python
# Plot count of cities by state
plt.figure(figsize=(12, 6))
sns.countplot(data=sushma, x='ship-state')
plt.xlabel('states')
plt.ylabel('count')
plt.title('Distribution of State')
plt.xticks(rotation=90)
plt.show()
```

# top-10

```python
top10= sushma['ship-state'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.countplot(data=sushma[sushma['ship-state'].isin(top10.index)], x='ship-state')
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of State')
plt.xticks(rotation=90)
plt.show()
```
